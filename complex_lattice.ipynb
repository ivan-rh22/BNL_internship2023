```python
"""complex_lattice.ipynb"""

# Cell 1

# Code is annotated version of solution provided by Dr. David Biersach

#IMPORTS
from __future__ import annotations

import typing
# plotting and other math tools
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.markers import MarkerStyle
from matplotlib.ticker import MultipleLocator

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from matplotlib.figure import Figure
    from numpy.typing import NDArray

%matplotlib widget
def f(z: NDArray[np.complex]) -> NDArray[np.complex_]:
    """
    Args:
        z: array of complex numbers
    Return:
        z^2 + z + 1j + 1 as an array of complex numbers
    """
    return z ** 2 + z * 1j + 1

x: NDArray[np.float_] = np.linspace(-4, 4, 1000)
y: NDArray[np.float_] = np.linspace(0, 2, 1000)

x, y = np.meshgrid(x, y)
z: NDArray[np.complex_] = f(x + y * 1j)

plt.close("all")
fig: Figure = plt.figure(" ")
ax: Axes = fig.add_subplot()
pixel: float = 72 / ax.figure.dpi ** 2
ax.scatter(np.real(z), np.imag(z), marker = MarkerStyle("o"), lw = 0, s = pixel)
ax.scatter(x, y, marker = MarkerStyle("o"), lw = 0, s = pixel)
ax.set_title("f(z)")
ax.set_xlabel("Re")
ax.set_ylabel("Im")
ax.grid()
ax.set_aspect("equal")
plt.show()
```

```python
# Cell 2
# Count number of complex lattice points
# z is a Gaussian integer and |Re(f(z))| and |Im(f(z))| <= 10 

num_points = 0

for x in range(-500, 500):
    for y in range(-500, 500):
        if abs(x ** 2 - y ** 2 - y + 1) <= 10 and abs(2 * x * y + x) <= 10:
            num_points += 1

print(f"There were {num_points:,} lattice points")
```
