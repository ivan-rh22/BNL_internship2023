```python
"""agnesi_witch.ipynb"""

# Cell 1 - Express f(x) as a power series over -1 < x < 1

# CODE OBTAINED FROM DR. BIERSACH

from __future__ import annotations

import typing

import matplotlib.pyplot as plt
import numpy as np

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget

def f(x: NDArray[np.float_]) -> NDArray[np.float_]:
    """function that expresses Maria Agnesi's simplified equation"""
    return 1 / (1 + x**2)

def p_f(x: float, num_terms: int) -> float:
    """function that expresses power series approximation """
    y: float = 0.0
    for n in range(num_terms):
        y += (-(x**2)) ** n
    return y

# creating figure and axis
fig, ax = plt.subplots()

# x-axis range
x: NDArray[np.float_] = np.linspace(-1.5, 1.5, 1_000)

# plot exact formula
ax.plot(x, f(x), label="Exact")

# plotting series approximation
v = np.vectorize(p_f, excluded=("num_terms"))
for n in range(2, 8):
    ax.plot(x, v(x, num_terms=n), label=f"{n} terms")

# Labeling plot graph
ax.set_title("Runge's Phenonmenon (Witch of Agnesi)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc="best")

ax.set_ylim(-2, 2)

# Styling
ax.axvline(x=-1, color="black", linestyle="--", linewidth=2)
ax.axvline(x=1, color="black", linestyle="--", linewidth=2)

# Disply the grid
ax.grid()
plt.show()

```
